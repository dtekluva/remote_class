```python
# bubble chart using matplotlib and pandas
# place legend outside plot

import pandas
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# %matplotlib inline

# pandas.set_option('max_columns', 10)

# put in path/filename (see csv data below)
# data can also be found in data section of world bank site
df = pandas.read_csv('data.csv')
df
plt.scatter(x=df['GDP Per Capita'], 
y=df['Lifespan'], 
s=df['Population']/1000, 
alpha=0.5, 
c=df['Bubble Color'])

# # chart title, axis labels
plt.title('GDP, Lifespan, Population (bubble size)')
plt.xlabel('gdp')
plt.ylabel('lifespan')

# # bubble labels
x,y = df['GDP Per Capita'], df['Lifespan']
for i, txt in enumerate(df['Territory']):
    plt.annotate(txt, (x[i]-5000, y[i]))
#     print(i, txt, x[i], y[i], df['Population'][i], df['Bubble Color'][i])

# # legend
territory_list = list(df['Territory'])
bubble_color_list = list(df['Bubble Color'])
l = []

for i in range(0,len(df.index)):
    l.append(mpatches.Patch(color=bubble_color_list[i], alpha=0.5, label=territory_list[i]))
    
plt.legend(handles=l, loc=(1.05,0))
df
# # csv data
# '''
# Territory,GDP Per Capita,Lifespan,Population,Bubble Color
# Qatar,127523,78.2,2569804,blue
# Luxembourg,105882,82,575747,pink
# Macau,104169,80.8,612167,red
# Singapore,87856,83.1,5622455,cyan
# Brunei,77441,77.7,423196,green
# United Arab Emirates,72419,77.1,9269612,yellow
# Ireland,68883,81.4,4726078,black
# Switzerland,62882,83.4,8401739,gray
# Norway,59302,81.8,5254694,purple
# Hong Kong,58553,84.3,7302843,orange'''
```
