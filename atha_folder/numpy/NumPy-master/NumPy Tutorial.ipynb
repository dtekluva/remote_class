#### ASSIGNMENT 

###### FROM THE FILE SALES.CSV - 

- GET THE AVERAGE SALES FOR EACH WEEK IN THE DATASET, 
- ALSO GET THE TOTAL SALES FOR THE FIRST MONTH TH THE DATASET,
- GET THE DAY WITH THE HIGHEST SALES IN THE 3RD MONTH OF THE DATASET
- ON AVERAGE WHAT WEEK OF THE MONTH HAS THE HIGHEST SALES VOLUME.

** NOTE THAT - DATASET HAS MON-SUNDAY SALES FOR 115WEEKS (MEANING 2YEARS AND 2WEEKS DATA)

```python
filedata = np.genfromtxt('sales.csv', delimiter=',', )
filedata = filedata.astype('int16')
data = filedata[1:]
# print(data)
```

```python
weekly_average_sales = np.mean(data[:,1:], axis = 1)
for week, avg in (enumerate(weekly_average_sales)):
    print(f"Month {week}".center(10), f"- {avg}".ljust(10)  )
```

```python
first_month_sales = data[:4,1:]
first_month_sales.sum()
```

```python
# 3rd month highest sales week
days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

month3_weeks = data[4*2:4*3,1:].reshape(28)
max_value = month3_weeks.max()
occurence_index = np.where(month3_weeks == max_value)
day = (occurence_index[0] % 7)[0]

print(f"Max sales occurred on {days_of_week[day]} the {occurence_index[0][0]} day of the month.")
```

```python
# print(len(weekly_average_sales)) # we have 105 lines which is about 2years of data and 2extra weeks
#hence we will break it down into 2years of data and remove the 2 extra weeks

year1 = weekly_average_sales[:52]
year2 = weekly_average_sales[52:52*2]

year1_monthly = year1.reshape(13,4) 

year2_monthly = year1.reshape(13,4) 
year1_monthly
```

```python
def get_average(data):
    max_monthly = []
    
    for i in data:
        max_value = i.max()
        occurence_index = np.where(i == max_value)
        max_monthly.append(occurence_index[0][0])
        
    return max_monthly
        
cummulated_averages = get_average(year1_monthly) + get_average(year2_monthly)
```

```python
# Program to find most frequent
def most_frequent(_list): 
    dict = {} 
    count, itm = 0, '' 
    for item in reversed(_list): 
        dict[item] = dict.get(item, 0) + 1
        if dict[item] >= count : 
            count, itm = dict[item], item 
    return(itm) 
  
print(f"Most frequent high sales week is usually week {most_frequent(cummulated_averages)+1}.") 
```
